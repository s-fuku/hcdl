```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import pickle
```

```python
d = pd.read_csv('data_processed/FEH_200504_201403.csv')
```

```python
d.head(3)
```

```python
for j in range(4, d.shape[1]):
    d.iloc[:, j] = d.iloc[:, j].str.replace('-', '0').replace(',', '', regex=True).astype(np.int)
```

```python
d.head(3)
```

```python
ym_list = np.unique(d['Unnamed: 0'].values)
```

```python
ym_list
```

```python
d_list = []
for i, ym in enumerate(ym_list):
    pair_list = []
    idxes = np.where(d['Unnamed: 0'].values == ym)[0]
    d_ym = d.loc[idxes, '北海道':'沖縄県'].values
    
    for j in range(d_ym.shape[0]):
        for k in range(d_ym.shape[1]):
            #if d_ym[j, k] != 0:
            #    d_list.append([j+1, k+1, d_ym[j, k]])
            pair_list.append([j+1, k+1, d_ym[j, k]])
    np.savetxt('output/deltacon/edgelist_%03d.txt' % i, np.array(pair_list), fmt='%i')
    
    d_list.append(d_ym)
    
d_list_array = np.array(d_list)
with open('data_processed/X.pkl', 'wb') as f:
    pickle.dump(d_list_array, f)
```
